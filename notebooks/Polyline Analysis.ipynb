```python
err_x = err[err.columns[::2]]
err_y = err[err.columns[1::2]]
```

```python
dist = np.sqrt(err_x.as_matrix() ** 2 + err_y.as_matrix() ** 2).T
```

```python
fig, ax = plt.subplots()

threes = np.where(labels==0)
# _=ax[0][0].plot(dist[threes[0]].T)
plt.plot(dist[threes[0]].T.mean(axis=1)[18:85], label='0')
# ax[0][0].set_title('0')
# ax[0][0].set_ylim((-0.5,5))


threes = np.where(labels==1)
# _=ax[0][1].plot(dist[threes[0]].T)
plt.plot(dist[threes[0]].T.mean(axis=1)[18:85], label='1')
# ax[0][1].set_title('1')
# ax[0][1].set_ylim((-0.5,5))

threes = np.where(labels==2)
# _=ax[1][0].plot(dist[threes[0]].T)
plt.plot(dist[threes[0]].T.mean(axis=1)[18:85], label='2')
# ax[1][0].set_title('2')
# ax[1][0].set_ylim((-0.5,5))

threes = np.where(labels==3)
# _=ax[1][1].plot(dist[threes[0]].T)
plt.plot(dist[threes[0]].T.mean(axis=1)[18:85], label='3')
# ax[1][1].set_title('3')
# ax[1][1].set_ylim((-0.5,5))
fig.suptitle('Euclidean Distance between points on Poly410 and Poly470')
fig.tight_layout()
plt.subplots_adjust(top=0.85)
plt.legend()
fig.savefig('../figures/polylineDifference.pdf')
```
