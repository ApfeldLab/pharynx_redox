```python
import pandas as pd
import os
from glob import glob
```

```python
new_data_dir = "/Users/sean/code/wormAnalysis/data/NEW_DATA/"
for f in glob(new_data_dir + "**/*.csv"):
    print(f)
```

```python
big_df = pd.DataFrame()
sub_dfs = []
for f in glob(new_data_dir + "**/*.csv"):
    if 'indexer' in f:
        continue
    df = pd.DataFrame()
    df['movement'] = pd.read_csv(f, header=None, names=["movement"])['movement']
    if 'pair1' in f:
        df['pair'] = 1
    if 'pair2' in f:
        df['pair'] = 2
        
    if 'Anterior' in f:
        df['region'] = 'anterior'
    if 'Posterior' in f:
        df['region'] = 'posterior'
    if '_Tip' in f:
        df['region'] = 'tip'
    if 'SidesOfTip' in f:
        df['region'] = 'sides_of_tip'
    
    df['experiment'] = f.split('/')[-2]
    
#     print(df)
    sub_dfs.append(df)
    
big_df = pd.concat(sub_dfs)
print(big_df)
big_df.to_csv('~/Desktop/big_df.csv')
```
